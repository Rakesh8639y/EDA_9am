**covariance**

```python
x=range(1,11)
y=range(100,110)
len(x),len(y)
arr1=np.array(x)
arr2=np.array(y)
np.sum((arr1-np.mean(arr1))*(arr2-np.mean(arr2)))/9
```

**correlation**

```python
x1=arr1-np.mean(arr1)
y1=arr2-np.mean(arr2)
Sum=np.sum(x1*y1)
x2=np.sqrt(np.square(x1))
y2=np.sqrt(np.square(y1))
sum1=np.sum(x2*y2)
cor=Sum/sum1
cor   
```

**eludiean distance**

```python
x=(10,20)
y=(5,6)
arr1=np.array(x)
arr2=np.array(y)
np.sqrt(np.sum(np.square(arr1-arr2)))
```

**cosine similarity**

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
cos_sim=np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
cos_sim
```

**z_score**

```python
data=np.array([10,20,30,40,50,60])
z_score=(data-np.mean(data))/np.std(data)
z_score
```
